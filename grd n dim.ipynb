```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
```

```python
X,y = load_diabetes(return_X_y = True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = .2 ,random_state = 23)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
lr.coef_
```

```python
class NdimGRD:
    def __init__(self,lrate=.01,epochs=500):
        self.coef_ = None
        self.intercept_ = None
        self.epochs = epochs
        self.lrate = .1
    def fit(self,X_train,y_train):
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])

        for i in range(self.epochs):
            y_hat = np.dot(X_train,self.coef_) + self.intercept_
            inter_slope = -2 * np.mean(y_train - y_hat)
            self.intercept_ = self.intercept_  - (self.lrate * inter_slope)

            coef_slope = -2*np.dot((y_train-y_hat).T,X_train)/X_train.shape[0]
            self.coef_ = self.coef_ - (self.lrate * coef_slope)

    def predict(self,X_test):
        return np.dot(X_test,self.coef_)+self.intercept_
       
```

```python
r = NdimGRD()
```

```python
r.fit(X_train,y_train)
```

```python
r.intercept_
```

```python
r.coef_
```

```python
y_p = r.predict(X_test)
```

```python
r2_score(y_test,y_p)
```
