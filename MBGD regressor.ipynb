```python
import pandas as pd
from sklearn.datasets import load_diabetes
import numpy as np
import random
```

```python
X,y = load_diabetes(return_X_y = True)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2,random_state=23)
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
class MBGDRegressor:
    def __init__(self,batch_size = 10,lrate=.03,epochs=40):
        self.epochs = epochs
        self.lrate = lrate
        self.intercept_ = None
        self.coef_ = None
        self.batch_size = batch_size
    
    def fit(self,X_train,y_train):
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])

        for i in range(self.epochs):
            for j in range(int(X_train.shape[0]/self.batch_size)):
                idx = random.sample(range(X_train.shape[0]),self.batch_size)
                print(idx)
                y_hat = np.dot(X_train[idx],self.coef_)+self.intercept_

                intercept_slope = -2*np.mean(y_train[idx]-y_hat)
                self.intercept_ = self.intercept_ - (self.lrate*intercept_slope)

                coef_der = -2 * np.dot((y_train[idx]-y_hat),X_train[idx])
                self.coef_ = self.coef_ - (self.lrate*coef_der)
                
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

```python
clr = MBGDRegressor()
```

```python
clr.fit(X_train,y_train)
```

```python
y_pred_clr = clr.predict(X_test)
```

```python
clr.intercept_
```

```python
clr.coef_
```

```python
r2_score(y_test,y_pred_clr)
```
